#### Question 1.c) What are the features (columns) and how are they represented?
#### Answer 1.c) 
* The features are as listed below:
    * Country name -
       * DataType - object
       * Desc - Name of the country.
    * year -
       * Datatype - int64 
       * Desc - Year the data was collected from.
    * Life Ladder -
       * Datatype - float64 
       * Desc - Imagine a ladder with steps in a range of 0 to 10, with 0 being the worst and 10 being the best possible life.
    * Log GDP per capita -
       * Datatype - float64
       * Desc - GDP of the country for that particular year.
    * Social support -
       * Datatype - float64
       * Desc - Average of responses(0s or 1s) from the people, when asked whether they had a relative or a friend to help when they needed them in trouble. 
    * Healthy life expectancy at birth -
       * Datatype - float64
       * Desc - Mean of number of years a person is expected to live in full health at birth.
    * Freedom to make life choices -
       * Datatype - float64
       * Desc - Average of responses(0s or 1s) from the people, when asked whether they were satisfied with the freedom to choose what they want in their life.
    * Generosity -
       * Datatype - float64
       * Desc - Average of responses(0s or 1s) from the people, when asked whether they donated money to the charity in the last month.
    * Perceptions of corruption -
       * Datatype - float64
       * Desc - Average of responses(0s or 1s) from the people, when asked whether corruption is widespread throughout the government or not, or if it was widespread throughout the businesses or not.
    * Positive affect -
       * Datatype - float64
       * Desc - Average of responses(0s or 1s) from the people, when asked whether they laughed, enjoyed, or did some interesting           things.
    * Negative affect -
       * Datatype - float64
       * Desc - Average of responses (0s or 1s) from the people upon asking whether they were worried, sad, or angry.
(Reference: John F. Helliwell, Haifang Huang, Max Norton, Shun Wang and Leonard Gof. https://happiness-report.s3.amazonaws.com/2023/WHR+23_Statistical_Appendix.pdf. 1-2)
